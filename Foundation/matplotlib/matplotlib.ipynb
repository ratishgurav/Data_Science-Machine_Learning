Matplotlib example workflow

```python
# immport matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

#prepare data
x=[1,2,3,4]
y=[11,22,33,44]

#Setup plot
fig, ax=plt.subplots(figsize=(10,10))#width ,height

#plot the data
ax.plot(x,y)

#customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

#save & show the whole figure
fig.savefig("sample_plot.png")
```

2 options for subplots

```python
#option1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2)

#plot to each diff axis
ax1.plot(x,x/2);
ax2.scatter(x,np.sin(x));
ax3.bar(nutbutter.keys(),nutbutter.values());
ax4.hist(x);
```

```python
#option 2
fig,ax=plt.subplots(nrows=2,ncols=2)

#plot to each diff index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(x,np.sin(x));
ax[1,0].bar(nutbutter.keys(),nutbutter.values());
ax[1,1].hist(x);
```

```python
#on another dataset
heart=pd.read_csv("heart-disease.csv")
heart
```

```python
heart['age'].hist(bins=10);
```

```python
heart.plot.hist(figsize=(10,30),subplots=True);
```

```python
## which one should u use matplot or oo method?
#quick plots use pyplot
#plotting something advanced use OO
```

```python
over_50=heart['age']>50#will give that columns only
over_50
```

```python
over_50=heart[heart['age']>50]#will get the entire dataframe
over_50
```

```python
#pyplot method
over_50.plot(kind="scatter",x='age',y='chol',c='target',figsize=(10,6));
```

```python
#OO  method with pyplot
fix,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax);
#ax.set_xlim([45,100])#changes limitations of x_axis
```

```python
fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50['age'],
                   y=over_50['chol'],
                   c=over_50['target'])

#customize the plot
ax.set(title="heart disease and cholesterol levels",xlabel="age",ylabel="chol",
       );

#legend
ax.legend(*scatter.legend_elements(),title="Target");

#add a horizontal line
ax.axhline(over_50['chol'].mean(),linestyle="--")
```

```python
over_50
```

```python
#subbplot of chol,age,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

#plot ax0
scatter=ax0.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])

#customize ax0
ax0.set(title="Heart disease with cholesterol",ylabel="Chol");
#legend ax0
ax0.legend(*scatter.legend_elements(),title="Target");
#add horizontal line ax0
ax0.axhline(over_50['chol'].mean(),linestyle="--");

#plot ax1
scatter=ax1.scatter(x=over_50['age'],
                    y=over_50['thalach'],
                    c=over_50['target'])
#customize ax1
ax1.set(title="Heart disease with Thalach",xlabel="Age",ylabel="Thachal");
#legend ax1
ax1.legend(*scatter.legend_elements(),title="Target");
#add horizontal line to ax1
ax1.axhline(over_50['thalach'].mean(),linestyle="--");

#add title to figure
fig.suptitle("Heart disease Analysis",fontsize=16,fontweight="bold");
```

```python
#setting style
plt.style.use('seaborn-v0_8-whitegrid')

fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50['age'],
                   y=over_50['chol'],
                   c=over_50['target'],
                   cmap="winter")#this changees the color scheme

#customize the plot
ax.set(title="heart disease and cholesterol levels",xlabel="age",ylabel="chol",
       );

#legend
ax.legend(*scatter.legend_elements(),title="Target");

#add a horizontal line
ax.axhline(over_50['chol'].mean(),linestyle="--")
```

```python
#customizing y and x limitations
#subbplot of chol,age,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

#plot ax0
scatter=ax0.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target']
                    ,cmap="winter")

#customize ax0
ax0.set(title="Heart disease with cholesterol",ylabel="Chol");
#legend ax0
ax0.legend(*scatter.legend_elements(),title="Target");
#add horizontal line ax0
ax0.axhline(over_50['chol'].mean(),linestyle="--");

ax0.set_xlim([50,80])#changes both the ax
ax1.set_ylim([60,200])#only changes that particular ax

#plot ax1
scatter=ax1.scatter(x=over_50['age'],
                    y=over_50['thalach'],
                    c=over_50['target'],
                    cmap="plasma")
#customize ax1
ax1.set(title="Heart disease with Thalach",xlabel="Age",ylabel="Thachal");
#legend ax1
ax1.legend(*scatter.legend_elements(),title="Target");
#add horizontal line to ax1
ax1.axhline(over_50['thalach'].mean(),linestyle="--");

#add title to figure
fig.suptitle("Heart disease Analysis",fontsize=16,fontweight="bold");

```
