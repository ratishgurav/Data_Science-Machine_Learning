Dot product

```python
np.random.seed(0)
mat1=np.random.randint (10, size=(5,3))
mat2=np.random.randint (10, size=(5,3))
```

```python
mat1
```

```python
mat2
```

```python
#element wise multiplication
mat1*mat2
```

```python
#dot product
np.dot(mat1,mat2)
```

```python
#inside matrix number must match eg: (3x3) and (3x2) and resultant will be (3x2)
mat3=np.dot(mat1,mat2.T) #mat2(5,3) becomes (3,5) mat1=(5,3)
mat3
```

```python
mat3.shape
```

Dot product nutbutter example

```python
np.random.seed(0)
sales_amounts=np.random.randint(20,size=(5,3))
sales_amounts
```

```python
# create weekly sales dataframe
weekly_sales=pd.DataFrame(sales_amounts,index=["Mon","Tue","Wed","Thu","Fri"],
                          columns=["AlmondButter","PeanutButter",
                                   "CashewButter"])
weekly_sales
```

```python
#create prices array
prices=np.array([10,8,12])
prices
```

```python
#make dataframe
prices=pd.DataFrame(prices.reshape(1,3),index=["Price"],
                          columns=["AlmondButter","PeanutButter",
                                    "CashewButter"])
prices
```

```python
#take transpose
total_sales=prices.dot(weekly_sales.T)
total_sales
```

```python
weekly_sales["Total"]=total_sales.T
weekly_sales
```
