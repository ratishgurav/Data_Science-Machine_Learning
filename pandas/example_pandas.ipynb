Manipulating Data

```python
car_sales['Make']=car_sales['Make'].str.lower()
car_sales
```

```python
car_sales_missing=pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
car_sales_missing['Odometer'].fillna(car_sales_missing['Odometer'].mean(),inplace=True)
#or car_sales_missing=car_sales_missing['Odometer'].fillna(car_sales_missing['Odometer'].mean())
```

```python
car_sales_missing_dropped=car_sales_missing.dropna()
```

```python
car_sales_missing_dropped.to_csv=("car_sales_missing_dropped.csv")
```

```python
#column from series
seats_column=pd.Series([5,5,5,5,5])
car_sales['Seats']=seats_column
car_sales
car_sales['Seats'].fillna(5,inplace=True)
car_sales
```

```python
#column from list
fuel_economy=[7.5,9.2,5.0,8.7,9.8,7.9,7.6,6.9,10,8.8]
car_sales['fuel/km']=fuel_economy
car_sales
```

```python
car_sales['total fuel used']=car_sales['Odometer (KM)']/100*car_sales['fuel/km']

car_sales['total fuel (L)']=car_sales['Odometer (KM)']/100*car_sales['fuel/km']
car_sales
```

```python
#create column from a single value
car_sales['number of wheels']=4
car_sales
```

```python
car_sales['Passed road safety']=True
car_sales.dtypes
```

```python
#axis=1 is column and axis=2 is rows
car_sales.drop('total fuel used',axis=1,inplace=True)
car_sales

```

```python
# selects random small amouts of data
car_sales_shuffled=car_sales.sample(frac=1)
car_sales_shuffled
```

```python
car_sales_shuffled.reset_index(drop=True,inplace=True)
car_sales_shuffled
```

```python
car_sales['Odometer KM']=car_sales['Odometer (KM)'].apply(lambda x:x/1.6)
car_sales
```
